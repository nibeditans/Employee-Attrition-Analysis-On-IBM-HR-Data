<h3 style="color:blue;">Attrition Analysis</h3>

<b style="color:purple;">Find the percentage of employees who have left 
    <b style="color:green; background-color:yellow">(Attrition = "Yes")</b> vs. those who stayed
    <b style="color:green; background-color:yellow">(Attrition = "No")</b>.</b>

```python
left_count = df["Attrition"].value_counts()["Yes"]
stayed_count = df["Attrition"].value_counts()["No"]
total_emps = df.shape[0]
```

```python
left_per = round((left_count / total_emps) * 100, 2)
left_per
```

```python
stayed_per = round((stayed_count / total_emps) * 100, 2)
stayed_per
```

<b style="color:purple;">Create a count plot to visualize the Attrition distribution.</b>

```python
plt.figure(figsize = (8, 4))
sns.countplot(x = df["Attrition"], color = "g")
plt.title("Attrition Distribution")
plt.show()
```

<b style="color:purple;">Compare Average 
    <b style="color:green; background-color:yellow">Monthly Income</b> of employees who left vs. those who stayed.</b>

```python
left_df = df[df['Attrition'] == 'Yes']
stayed_df = df[df['Attrition'] == 'No']
```

```python
avg_income_left = left_df['MonthlyIncome'].mean().round(2)
avg_income_stayed = stayed_df['MonthlyIncome'].mean().round(2)
```

```python
avg_income_left
```

```python
avg_income_stayed
```

```python
income_diff = avg_income_stayed - avg_income_left
income_diff.round(2)
```
